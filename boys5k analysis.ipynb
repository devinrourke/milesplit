```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('boys5k.csv')
df.describe()
plt.scatter(df['Year'], df['Time'], s=1, c='k', marker='o')
axes = plt.gca()
axes.set_xlim([1999,2017])
axes.set_ylim([500,3500])
plt.show()
```

```python
x_total = df.groupby(['Year']).mean()['Time']
print x_total
print type(x_total)
x_total.plot()
plt.axis([1998, 2018, 1050, 1200])
plt.ylabel('seconds')
plt.title('nationwide average 5k time, boys')
plt.show()
```

```python
x_state = df.groupby(['State','Year']).mean()['Time']

ind = 0
for key in sorted(df['Time'].groupby([df['State']]).groups.keys()):    
    ind = ind + 1 
    plt.subplot(6, 10, ind)
    plt.xticks([])
    frame1 = plt.gca()
    frame1.axes.get_xaxis().set_visible(False)
    x_state[key].fillna(0).plot(title = key, figsize = (16, 9),
                                yticks = [900, 1150, 1400], fontsize = 12, sharey = True, sharex = False)
    plt.axis([1998, 2018, 850, 1450])
    plt.tight_layout()
    plt.xlabel('')
plt.show()
```

# Completeness of the data

```python
years = ['2000','2001','2002','2003','2004','2005','2006','2007',
         '2008','2009','2010','2011','2012','2013','2014','2015','2016']
years = map(int,years)

ind = 0
for key in sorted(df['Time'].groupby([df['State']]).groups.keys()):    
    ind = ind + 1
    ax = plt.subplot(6, 10, ind)
    plt.axis([2000, 2016, 0, 1000])
    plt.xticks([])
    frame1 = plt.gca()
    frame1.axes.get_xaxis().set_visible(False)
    df['Time'].groupby([df['State'], df['Year']]).count().get(key).reindex(index=years).plot.bar(
        title = key, figsize = (16, 9), fontsize = 12, sharey = True, sharex = False) 
    plt.tight_layout()
    plt.xlabel('')
plt.show()
```
